```python
#Plotting 3D graph
```

```python
x=np.random.randint(70,size=20)
y=np.random.randint(70,size=20)
z=np.random.randint(70,size=20)
```

```python
fig=plt.figure()
a=fig.add_subplot(projection='3d')
a.scatter(x,y,z)
plt.show()
```

```python
fig=plt.figure()
a=fig.add_subplot(projection='3d')
a.scatter(x,y,z,c='g',alpha=.8)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.show()
```
